```python
import torch
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/last.pt', force_reload=True)
```

```python
%cd /content
```

```python
# Image 1

!wget https://github.com/MuhammadHananAsghar/Thermal-Imaging-Object-Detection/raw/main/images/1.jpg
```

```python
image = cv2.imread("1.jpg")
normed = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)
```

```python
cv2_imshow(image)
```

```python
# Predictions

result = model(image)
result.print()

cv2_imshow(np.squeeze(result.render()))
```
